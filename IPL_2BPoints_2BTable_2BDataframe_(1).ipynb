Question-2:
If all the stats are taken for both ‘ipl17’ and ‘ipl18’, which team with its total points greater than 25 will have the highest win percentage?

QUESTION-1

```python
q1= ipl18[ipl18['NRR']>0]
q11 = q1.query('For > Against')
q11
```

QUESTION-2

```python
ipl = pd.merge(left=ipl17,right=ipl18,how='outer',left_on='Team',right_on='Team')
ipl
```

```python
ipl1 = ipl.copy()
```

```python
ipl1.fillna(0,inplace=True)
```

```python
ipl1['Total_Points']=ipl1['Points_x']+ipl1['Points_y']
ipl1['Total_Matches']=ipl1['Matches_x']+ipl1['Matches_y']
ipl1['Total_won']=ipl1['Won_x']+ipl1['Won_y']
ipl1['Total_loss']=ipl1['Lost_x']+ipl1['Lost_y']
ipl_fin = ipl1[['Team','Total_Points','Total_Matches','Total_won','Total_loss']]
ipl_fin= ipl_fin[ipl_fin['Total_Points']>25]
```

```python
q2 = ipl_fin.copy()
q2['Win %']= q2['Total_won']/q2['Total_Matches']*100
q2
```

```python
q2_fin = q2[['Team','Total_Points','Win %']]
```

```python
q2_fin
```
